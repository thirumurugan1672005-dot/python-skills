## startswith and endswith
* startswith : startswith with particular string
* endswith : checks with ends with the string

```python
# startswith or endswith
sentence = "harry is working in the shop"
print(sentence.startswith("harry"))
print(sentence.endswith("shop"))
```

# find
* find : the finds the partiuclar substrings index if not return -1 (left to right)
* rfind : finds from right to left

```python
#find or index
name = "Hippo"
print(name.find("i"))
print(name.find("p"))
print(name.rfind("p"))
```

## index
*index : index of the substring (left to right) if not found throws value Error
* rindex : checks from right to left

```python
name = "ippi"
print(name.index("p"))
print(name.rindex("p"))
name.index("f")
```

## casefold
casefold makes to compare strings

```python
text = "english is one of the LANGAUAge"
text3 = "english is one of the langAuage"
print(text.casefold())
print(text3.casefold())
```

## Encode
Encode is used to encode the string

```python
name = "Master"
print(name.encode("UTF-16",errors="strict"))
print(name.encode("UTF-8",errors="strict"))
```

## expand tabs

## removeprefix and removesuffix 
* removeprefix : removes prefix from string
* removesuffix : remove suffix from string

```python
text = "the world is good for nothing"
print(text.removesuffix("ing"))
print(text.removeprefix("the"))
```

# String Formatting
String Formatting is the format strings in specific way

```python
name = "The %s %i " % ("name",32)
print(name)
```

```python
name = "{} is working in {}"
name.format("Hella","Kitchen")
```

```python
name = "{subject} is doing {action}"
name.format(subject="master",action="swimming")
```

```python
# f strings
age = 20
name = "Miller"
print(f"{name} is {age} years old")
```
