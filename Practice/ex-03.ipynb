# Conditional Statement
Conditional Statement is used to control the flow of statements using conditions

## if statement
* if statement is used to condtion
* indentation is used as  scope
* it is four spaces for indentation as per convention 

```python
age = int(input("Enter your age"))
print(f"your age is {age}")
if age >= 18:
    print("You are adult")
```

## else statement 
else statement is used to do when all the if conditions are false in if-elif-else ladder

```python
age = int(input("Enter your age: "))
print(f"your age is {age}")
if age >= 18:
    print("adult")
else:
    print("young")
```

## elif Statement
* elif Statement is used to given conditions if or elfi before it does not works 
* if-elif-else ladder works in top-bottom approach

```python
marks = int(input("Enter marks: "))
if marks < 0:
    print("Invalid marks!")
elif marks <= 35:
    print("F")
elif marks <= 45:
    print("E")
elif marks <= 55:
    print("D")
elif marks <= 65:
    print("C")
elif marks <= 75:
    print("B")
elif marks <= 85:
    print("A")
elif marks <= 90:
    print("A+")
else:
    print("O")
```

## Nested Conditions 
Nesting if  statements wiuthin another if scope

```python
n = int(input("Enter number:"))
print(f"num is {n}")
sen = ""
if n % 3 == 0:
    sen+="three"
if n % 5 == 0:
    sen+="five" 
print(sen)
```

## Ternary Operator
Ternary Operator is used to chose values based on condition

```python
n = int(input("Enter number:"))
is_even =  "Even" if n % 2 == 0 else "Odd"
print(is_even)
```

# Match Case Statement
Match case statement is used like switch statements to selec t like switch statements

```python
season = int(input("Enter season code :"))
match season:
    case 1 : 
        print("January")
    case 2:
        print("February")
    case 3:
        print("March")
    case 4:
        print("April")
    case 5:
        print("May")
    case 6:
        print("June")
    case 7:
        print("July")
    case 8 :
        print("August")
    case 9:
        print("September")
    case 10:
        print("October")
    case 11:
        print("November")
    case 12:
        print("December")
    case _ :
        print("Invalid month")
```

```python
#matching Integer
position = 15
match position:
    case 1:
        print("Pitcher")
    case 2:
        print("Catcher")
    case 3:
        print("First baseman")
    case 4:
        print("Second baseman")
    case 5:
        print("Third baseman")
    case 6:
        print("Shortstop")
    case 7:
        print("Left fielder")
    case 8:
        print("Center fielder")
    case 9:
        print("Right fielder")
    case _:
        print("Unknown position")
```

```python
# matching strings
minister = input("Enter minister name: ")
match minister:
    case "Vijay":
        print("Vijay is rulling")
    case "Keerthana":
        print("Keerthana is taking reels")
    case "Nirmal":
        print("Nirmal is showing his wealth")
    case _ :
        print("People are normal!")
```

```python
# match with if 
current = 500
money = 0
match current:
    case x if x > 0 and x <= 500:
        money = x - 200
    case x if x > 500:
        money = x-100
    case _ :
        print("Default option")
print(money)
```
