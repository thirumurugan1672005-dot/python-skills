# Looping Statements
Looping Statements are used to iterate over certain number of times to control the flow of statements

## While loop
While loop will execute until the condition becomes false 

```python
i = 0
while i < 5:
    print(i,end =" ")
    i+=1
```

## Range() function
* Range() function helps to  range of numbers start = 0 , end = n from 0 to n-1 
* start is optional
* step which is default by 1
* step which tells the what to add with current iterable

```python
range(12)
```

```python
range(1,23)
```

## for Loop
for loop is used to iterate over indexes or over range of numbers . definite iterations

```python
for i in range(12):
    print(i,end = " ")
```

```python
for i in range(2,12):
    print(i,end = " ")
```

```python
for i in range(2,12,2):
    print(i,end = " ")
```

```python
for i in range(12,2,-2):
    print(i,end = " ")
```

```python
for i in range(1,34,-2):
    print(i,end =" ")
```

```python
for i in range(34,-1,-2):
    print(i,end = " ")
```

## break statement
Break statement is used to stop the iteration
## continue statement
Continue Statement is used to skip the current iteration
## pass statement
Pass Statement is used to the when nothing in body to affect syntax Error

```python
for i in range(12):
    if i == 1:
        continue
    print(i,end = " ")
print()
for i in range(12):
    if i == 2:
        break
    print(i,end = " ")
while False:
    pass
```

## else clause
else clause is used to execute when loops finish without break statement

```python
pos=2
while pos < 10:
    if pos == 5:
        break
    print(pos,end = " ")
    pos+=1
else:
    print("while loop ends")
```

```python
pos = 2
while pos < 5:
    print(pos,end = " ")
    pos+=1
else:
    print("while loop ends sucessfully")
```

```python
for i in range(1,20):
    if i == 8:
        break
    print(i,end = " ")
else:
    print("loop executes")
print()
for i in range(1,5):
    print(i,end = " ")
else:
    print("loop executed")
```
