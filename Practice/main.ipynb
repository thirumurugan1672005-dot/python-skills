## Indexing
* Strings are  indexed from 0 to length of string -1

```python
name = "Intelligent" 
print(name[0],name[3])
print(name[-1],name[-3])
```

## Slicing
* Slicing is the feautre to take substrings from strings
* [start : stop : end] 
#### when end is positive default 1
* start : default value 0
* end : default value length of string
* end index was always excluded

```python
name = "Intelligent"
print(name[:])
print(name[3:])
print(name[:5])
print(name[2:4])
```

#### when end is  negative
1. start : -1 
2. end : -len(string)

```python
name = "Intelligent"
print(name[::-1])
print(name[2::-1])
print(name[:3:-1])
```

```python
name = "SanjayKumar" 
print(name[-5:-2])
```

# Logical Operators
* Logical Operators are used to take logical conditions
* and : If Both of them True it gives true
* or : If Either of themTrue it is True
* not Flips the value

```python
# check age >= 18 
age = int(input("Enter the number: "))
is_valid = True
print(f"The given age is {age}")
age >= 18 and is_valid
```

```python
# check the leap or not
year = int(input("Enter the year : "))
print("Entered year: ",year)
is_leap = year % 400 == 0 or (year % 100 != 0 and year % 4 == 0)
print(is_leap)
```

```python
# not even 
num = int(input("Enter the number : "))
is_not_even = not num % 2 == 0
print(is_not_even)
```
